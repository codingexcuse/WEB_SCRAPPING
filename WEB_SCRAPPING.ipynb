```python
# This is to get information from an html file..
with open("formss.html","r") as html_file:
    content = html_file.read()
    print(content)
```

```python
# print the html code using beautifulsoup..
soup = BeautifulSoup(content,"lxml")
print(soup.prettify())
```

```python
# Get all the h1 tags are in part of code..
tags  = soup.find('h1')
print(tags)

# the difference betweeen find and all_find is find stop for the first one itself but where as find_all will find everything..
all_tags = soup.find_all('h1')
for i in all_tags:
    print(i.text)
```

```python
# get the companies who are offering the jobs for python..
html_info = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=')
soup = BeautifulSoup(html_info.text,"lxml")
jobs = soup.find_all('li',class_ = 'clearfix job-bx wht-shd-bx')
skills = soup.find_all('span',class_ = "srp-skills")
published_date = soup.find_all('span',class_ = "sim-posted")
company_jobs = []
company_skills = []
company_published = []
for i in jobs:
    company_jobs.append(i.h3.text.replace(" ",""))
for i in skills:
    company_skills.append(i.text.replace(" ",""))
for i in published_date:
    company_published.append(i.text.replace(" ",''))
```

```python
# how to remove \n and \r from the list...

#for i in company_jobs:
#    print(i.rstrip())

        # OR..
company_jobss = list(map(str.rstrip,company_jobs))
for i in company_jobss:
    print(i)
print(len(company_jobss))
```

```python
for i in range(0,25):
    print("\x1B[4m" + "company:" + "\x1B[0m"+company_jobs[i].rstrip())
    print()
    print("\x1B[4m" + "published_date:" + "\x1B[0m" +company_published[i].rstrip())
    print()
    print("\x1B[4m" + "skill:" + "\x1B[0m" +company_skills[i].rstrip())
    print()
    print("------------------------------------")
```

```python
html_info = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=')
soup = BeautifulSoup(html_info.text,"lxml")

unskill = input(">")

jobs = soup.find_all('li',class_ = "clearfix job-bx wht-shd-bx")

name = []
for i in jobs:
    company_name = i.find('h3',class_ = "joblist-comp-name").text.replace(" ","")
    company_skill = i.find('span',class_ = "srp-skills").text.replace(" ","")
    company_date = i.find('span',class_ = "sim-posted").text.replace(" ","")
    company_link = i.header.h2.a["href"]
    if unskill not in company_skill:
        name.append(company_name)
        print(f"company_name : {company_name} company_skill : {company_skill} company_date  : {company_date} More_Info : {company_link}") 
        print("________________________________________________________________________________________________________")
```
