# IMDB Ratings...... :)


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# print the most high rated horror films in IMDB...

html_info = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=horror&sort=user_rating,desc")
soup = BeautifulSoup(html_info.text,"lxml")

films = soup.find_all("div",class_ = "lister-item mode-advanced")



movie__name = []
movie_year = []
movie_rating = []
movie_gross = []
movie_votes = []




for i in films:
    movie_name = i.h3.a.text
    movie__name.append(movie_name)
    year = i.find("span",class_ = "lister-item-year text-muted unbold").text
    movie_year.append(year)
    rating = i.find("div",class_ = "inline-block ratings-imdb-rating").text.replace("\n","")
    movie_rating.append(rating)
    gross_and_votes = i.find_all("span",attrs = {"name" : "nv"})
    
    votes = gross_and_votes[0].text
    movie_votes.append(votes)
    gross = gross_and_votes[1].text if len(gross_and_votes) >1 else "NA"
    movie_gross.append(gross)
    
    
    print("\x1B[4m" + "movie name :" +"\x1B[0m" + " "+movie_name )
    print("\x1B[4m" + "year of released :" +"\x1B[0m"+ " "+year)
    print("\x1B[4m" + "rating :" +"\x1B[0m"+" "+str(rating))
    print("\x1B[4m" + "Gross of Movie :" +"\x1B[0m"+ " "+str(gross))
    print("\x1B[4m" + "votes :" +"\x1B[0m"+ " "+votes)
    
    
    print("------------------------------------")
    print()
```

```python
# Now create a data frame and load the data from the above movies data...
movie_df = pd.DataFrame({"movie " : movie__name, "year " : movie_year, "Rating " : movie_rating, "votes " : movie_votes, "gross ": movie_gross})
```
